# **Project Summary -**

This project focuses on performing a comprehensive **Exploratory Data Analysis (EDA)** of Netflix’s catalog of movies and TV shows to uncover meaningful insights about the platform’s content strategy, audience preferences, and business direction.  
The dataset used contains **7,787 records and 12 columns**, providing information such as title, type (Movie or TV Show), director, cast, country, date added, release year, rating, duration, genres (listed_in), and description. The data was sourced from **Flixable**, a third-party Netflix search engine, and represents titles available on Netflix up to 2019.

In addition to EDA, this project serves as the foundation for a **machine learning (ML) task using K-Means clustering**, where the goal is to group similar titles based on features such as genre, rating, duration, and release period.  
The insights derived from EDA help define relevant features, guide preprocessing steps, and ensure the data is well-prepared for clustering analysis.

---

### **1. Data Cleaning and Preprocessing**

The initial step involved inspecting and cleaning the dataset. Columns such as `director`, `cast`, `country`, and `rating` contained missing values, which were replaced with `"Unknown"` to retain records without losing data.  
The `date_added` column was converted into a proper datetime format, and two new columns — `year_added` and `month_added` — were derived for trend analysis.  
The `duration` column was further split into `duration_minutes` (for movies) and `seasons` (for TV shows) to enable quantitative analysis across both types of content.  
The dataset was found to be free of duplicates, with minimal data inconsistencies, making it ready for both visual exploration and machine learning applications.

---

### **2. Exploratory Data Analysis (EDA)**

The EDA process involved generating a series of visualizations to interpret Netflix’s content trends:

* **Content Distribution:** Movies dominate the catalog with around 70% share, while TV Shows make up the remaining 30%. However, the number of TV Shows has increased significantly after 2015, showing Netflix’s growing investment in serialized content.  
* **Yearly and Monthly Trends:** Most titles were added to Netflix between 2016–2019, indicating aggressive content expansion during that period. Content additions are evenly spread across months, ensuring consistent user engagement throughout the year.  
* **Regional Contributions:** The United States leads with the highest number of titles, followed by India, the United Kingdom, and Japan. A deeper comparison between **India and the US** showed that while the US dominates in volume, India has shown a steep growth trajectory since 2016 — reflecting Netflix’s focus on international markets and localization.  
* **Ratings:** Ratings analysis revealed that most titles are rated **TV-MA** and **TV-14**, implying a heavy focus on adult and teenage audiences. Family-friendly content such as **PG** or **TV-Y** is relatively limited, representing an opportunity for expansion.  
* **Duration and Seasons:** Most movies are between **80–120 minutes**, and most TV shows have **1–2 seasons**, aligning with Netflix’s binge-watch culture.  
* **Genres:** Popular genres include **International Movies, Dramas, Comedies, and Documentaries**, showing a preference for storytelling variety and global appeal.  
* **Directors and Cast:** A few directors, like *Raúl Campos, Jan Suter,* and *Marcus Raboy*, contribute multiple titles, often in stand-up or documentary categories, highlighting Netflix’s collaboration with consistent creators.

---

### **3. Numerical Analysis**

This diversity demonstrates Netflix’s strategy of providing a balanced range of content to appeal to varied audiences, and also ensures feature independence — an ideal property for unsupervised learning models like **K-Means clustering**.

````python
plt.figure(figsize=(10,6))

# Create a new column: difference between year_added and release_year
data['content_gap'] = data['year_added'] - data['release_year']

# Filter out unrealistic gaps (negative or >20 years just in case of wrong entries)
data_gap = data[(data['content_gap'] >= 0) & (data['content_gap'] <= 20)]

# Plot the average gap per release year
avg_gap = data_gap.groupby('release_year')['content_gap'].mean().reset_index()

ax = sns.lineplot(x='release_year', y='content_gap', data=avg_gap, marker='o')
plt.title("Average Delay (Years) Between Release and Addition to Netflix")
plt.xlabel("Release Year")
plt.ylabel("Average Gap (Years)")
plt.tight_layout()
plt.show()
````

### 2. Handling Outliers

````python
# List of numeric columns to check for outliers
num_cols = ['duration_minutes', 'release_year', 'content_gap']

# Function to remove outliers using IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    # Filter data within range
    return df[(df[column] >= lower_limit) & (df[column] <= upper_limit)]

# Apply outlier removal for each selected column
print("Shape before removing outliers:", data.shape)
for col in num_cols:
    before = data.shape[0]
    data = remove_outliers_iqr(data, col)
    after = data.shape[0]
    print(f"{col}: removed {before - after} rows")

print("Shape after removing outliers:", data.shape)
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

To handle outliers in my dataset, I used the IQR (Interquartile Range) method.
This technique calculates the range between the first (Q1) and third (Q3) quartiles and removes any values that fall below Q1 − 1.5×IQR or above Q3 + 1.5×IQR.
I applied this method to numerical columns such as duration_minutes_est, release_year, and content_gap.

I chose the IQR method because it is simple, effective, and works well for datasets that are not normally distributed.
It helps to detect extreme values without being influenced by skewness or a few large numbers.
By removing these outliers, I ensured that the data used for clustering was clean, balanced, and not distorted by extreme or unrealistic values.

### 4. Textual Data Preprocessing

#### 1. Expand Contraction

````python
import contractions


# Select all object (text) columns
text_columns = data.select_dtypes(include=['object']).columns
print(" Text columns found:", list(text_columns))

# Safe function to expand contractions
def safe_expand(text):
    try:
        # Skip if not a valid string
        if not isinstance(text, str) or text.strip() == '':
            return text
        return contractions.fix(text)
    except Exception:
        return text  # If any weird text causes error, return it as is

# Apply safely to all text columns
for col in text_columns:
    data[col] = data[col].apply(safe_expand)

print(" Contractions expanded in all text columns safely!")
````

#### 6. Rephrase Text

````python
import re
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download('punkt')

# Select text columns
text_columns = data.select_dtypes(include=['object']).columns
print("Text columns found:", list(text_columns))

# Simple function to rephrase text
def simple_rephrase(text):
    try:
        if isinstance(text, str):
            # Split into sentences
            sentences = sent_tokenize(text)
            rephrased_sentences = []
            for sent in sentences:
                # Remove duplicate words like "very very" → "very"
                sent = re.sub(r'\b(\w+)( \1\b)+', r'\1', sent)
                # Remove extra spaces
                sent = re.sub(r'\s+', ' ', sent)
                # Optional: Fix common patterns like "it's about" → "the story is about"
                sent = sent.replace("it's about", "the story is about")
                sent = sent.replace("its about", "the story is about")
                rephrased_sentences.append(sent.strip())
            return " ".join(rephrased_sentences)
        return text
    except Exception:
        return text

# Apply rephrasing to each text column
for col in text_columns:
    data[col] = data[col].apply(simple_rephrase)

print("Text rephrasing (basic cleanup) completed successfully!")
````

#### 8. Text Normalization

````python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

# Download required NLTK resources
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')

# Lemmatizer initialization
lemmatizer = WordNetLemmatizer()

# Function to get POS (part of speech) tag for better lemmatization
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ, "N": wordnet.NOUN, "V": wordnet.VERB, "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

# Function to normalize text (works on tokenized lists)
def normalize_text(tokens):
    try:
        if isinstance(tokens, list):
            lemmatized = [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in tokens]
            return lemmatized
        elif isinstance(tokens, str):
            # If not tokenized, split and normalize
            words = tokens.split()
            lemmatized = [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in words]
            return lemmatized
        return tokens
    except Exception:
        return tokens

# Apply normalization to the tokenized text columns
text_cols_for_normalization = ['title_tokens', 'description_tokens']

for col in text_cols_for_normalization:
    data[col + '_normalized'] = data[col].apply(normalize_text)

print("Text normalization (lemmatization) completed successfully!")

# Example
print("\n Example:")
print("Before:", data['description_tokens'].iloc[0])
print("After Normalization:", data['description_tokens_normalized'].iloc[0])
````

##### Which text normalization technique have you used and why?

For text normalization, I used the Lemmatization technique provided by NLTK’s WordNetLemmatizer.
Lemmatization reduces each word to its base or dictionary form (known as the lemma) while considering the part of speech (POS) of the word.
For example, words like “playing”, “played”, and “plays” are all converted to “play”, and “children” becomes “child”.

I chose lemmatization over stemming because it is more accurate and context-aware — it doesn’t just cut off word endings but ensures that each word is linguistically correct.
This technique helps in reducing redundant variations of the same word and makes the text data more consistent and meaningful for machine learning models like clustering or sentiment analysis.

````python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from scipy.stats import skew, kurtosis


# Reduce Feature Correlation using PCA


print("Performing PCA to minimize correlated TF-IDF features...")

# Scale the data before PCA
scaler = StandardScaler(with_mean=False)  # with_mean=False for sparse data like TF-IDF
tfidf_scaled = scaler.fit_transform(tfidf_df)

# Apply PCA to reduce dimensionality (retain 95% variance)
pca = PCA(n_components=0.95, random_state=42)
tfidf_pca = pca.fit_transform(tfidf_scaled)

print(f"Reduced from {tfidf_df.shape[1]} → {tfidf_pca.shape[1]} principal components.")

# Convert PCA components back into a DataFrame
pca_df = pd.DataFrame(tfidf_pca, columns=[f'PC{i+1}' for i in range(tfidf_pca.shape[1])])

# Merge PCA-reduced features with original Netflix dataset
netflix_reduced_df = pd.concat([data.reset_index(drop=True), pca_df], axis=1)


#  Check Correlation Matrix (Post-PCA)

corr_matrix = pd.DataFrame(pca_df).corr()
high_corr = np.sum(np.abs(corr_matrix.values[np.triu_indices_from(corr_matrix, k=1)]) > 0.8)

print(f"Correlation check done — {high_corr} highly correlated pairs remaining (should be near 0).")


#  Create New Engineered Features


print("Creating new engineered features...")

# 🎬 Text-based Features
netflix_reduced_df['Description_Length'] = data['description'].str.split().apply(len)
netflix_reduced_df['Unique_Words'] = data['description'].apply(lambda x: len(set(str(x).split())))
netflix_reduced_df['Lexical_Richness'] = netflix_reduced_df['Unique_Words'] / netflix_reduced_df['Description_Length']

# Temporal Features
if 'release_year' in data.columns:
    netflix_reduced_df['Is_Recent'] = data['release_year'].apply(lambda x: 1 if x >= 2018 else 0)
    netflix_reduced_df['Title_Age'] = 2025 - data['release_year']

# Category/Genre Features
if 'listed_in' in data.columns:
    netflix_reduced_df['Num_Genres'] = data['listed_in'].apply(lambda x: len(str(x).split(',')))

# Statistical Features from PCA Components
netflix_reduced_df['PCA_Mean'] = pca_df.mean(axis=1)
netflix_reduced_df['PCA_Std'] = pca_df.std(axis=1)
netflix_reduced_df['PCA_Skew'] = pca_df.apply(lambda x: skew(x), axis=1)
netflix_reduced_df['PCA_Kurtosis'] = pca_df.apply(lambda x: kurtosis(x), axis=1)

# ------------------------------------------------
# 4️⃣ Final Feature Overview

print("\nFeature Manipulation Completed!\n")
print(f"Final dataset shape: {netflix_reduced_df.shape}")
print("\nNewly added features:")
print([
    'Description_Length', 'Unique_Words', 'Lexical_Richness',
    'Is_Recent', 'Title_Age', 'Num_Genres',
    'PCA_Mean', 'PCA_Std', 'PCA_Skew', 'PCA_Kurtosis'
])


#  Save for Next ML Steps

netflix_reduced_df.to_csv("Netflix_Feature_Engineered.csv", index=False)
print("\n Saved 'Netflix_Feature_Engineered.csv' successfully!")
````

````python
# FEATURE SELECTION TO AVOID OVERFITTING

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import VarianceThreshold, SelectKBest, mutual_info_classif, RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score



# 1. Separate Features and Target

target_col = 'Cluster'  # adjust if needed
X = data.drop(columns=[target_col], errors='ignore')
y = data[target_col]

# Drop non-numeric features
X = X.select_dtypes(include=[np.number])

print("Initial shape of X:", X.shape)


# 2. Remove Near-Zero Variance Features

var_thresh = VarianceThreshold(threshold=0.01)
X_var = X.loc[:, var_thresh.fit(X).get_support()]
print("After variance threshold:", X_var.shape)


# 3. Remove Highly Correlated Features

corr_matrix = X_var.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
high_corr_features = [col for col in upper.columns if any(upper[col] > 0.85)]
X_corr = X_var.drop(columns=high_corr_features)
print("Removed correlated features:", len(high_corr_features))
print("After correlation filter:", X_corr.shape)


# 4. Scale Features

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_corr), columns=X_corr.columns)


# 5. Mutual Information Selection

mi_selector = SelectKBest(score_func=mutual_info_classif, k=100)
mi_selector.fit(X_scaled, y)
mi_features = X_scaled.columns[mi_selector.get_support()]
X_mi = X_scaled[mi_features]
print("After Mutual Information Selection:", X_mi.shape)


# 6. Model-Based Recursive Feature Elimination (RFE)

log_reg = LogisticRegression(max_iter=1000, random_state=42)
rfe_selector = RFE(log_reg, n_features_to_select=50)
rfe_selector.fit(X_mi, y)

rfe_features = X_mi.columns[rfe_selector.get_support()]
X_rfe = X_mi[rfe_features]
print("After RFE Selection:", X_rfe.shape)


# 7. Stability Check Using Random Forest

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_rfe, y)
importances = pd.Series(rf.feature_importances_, index=X_rfe.columns)
selected_features = importances.sort_values(ascending=False).head(30).index

X_final = X_rfe[selected_features]
print("Final selected feature count:", X_final.shape[1])


# 8. Evaluate for Overfitting Risk

X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42, stratify=y)

rf_eval = RandomForestClassifier(n_estimators=100, random_state=42)
rf_eval.fit(X_train, y_train)

train_acc = accuracy_score(y_train, rf_eval.predict(X_train))
test_acc = accuracy_score(y_test, rf_eval.predict(X_test))

print("\nModel evaluation to check overfitting:")
print("Train Accuracy:", round(train_acc, 4))
print("Test Accuracy :", round(test_acc, 4))
print("Generalization Gap:", round(train_acc - test_acc, 4))


# 9. Save Final Selected Features

final_data = pd.concat([X_final, y.reset_index(drop=True)], axis=1)
final_data.to_csv("Netflix_Feature_Selected_Final.csv", index=False)
print("\nFeature selection completed and saved as 'Netflix_Feature_Selected_Final.csv'")
````

````python
import pandas as pd
from imblearn.over_sampling import SMOTE
from collections import Counter
from sklearn.preprocessing import LabelEncoder

# Load your data
X_train = pd.read_csv("Netflix_X_train.csv")
y_train = pd.read_csv("Netflix_y_train.csv")

# Convert DataFrame to Series if needed
if isinstance(y_train, pd.DataFrame):
    y_train = y_train.squeeze()

# Convert continuous cluster values back to discrete labels
# Example: map unique continuous values to integers (0,1,2,3,4)
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)

print("Before SMOTE:")
print(Counter(y_train_encoded))

# Apply SMOTE
smote = SMOTE(random_state=42, k_neighbors=3)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train_encoded)

print("\nAfter SMOTE:")
print(Counter(y_train_res))

# Optional: convert encoded labels back to original cluster values
y_train_res = encoder.inverse_transform(y_train_res)

# Save balanced dataset
X_train_res.to_csv("Netflix_X_train_balanced.csv", index=False)
pd.Series(y_train_res, name='Cluster').to_csv("Netflix_y_train_balanced.csv", index=False)

print("\nImbalanced dataset handled successfully.")
````

To handle the imbalance in the dataset, I used SMOTE (Synthetic Minority Oversampling Technique).
It balances the dataset by generating synthetic samples for minority clusters instead of removing data from majority clusters.
I chose SMOTE because it preserves all original data, reduces model bias toward majority classes, and improves the model’s ability to learn patterns from underrepresented clusters.

````python
# MODEL 1 - LOGISTIC REGRESSION
# CROSS VALIDATION & HYPERPARAMETER TUNING


import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# ------------------------------------------------
# Load Data
# ------------------------------------------------
X_train = pd.read_csv("Netflix_X_train_balanced.csv")
y_train = pd.read_csv("Netflix_y_train_balanced.csv").squeeze()
X_test = pd.read_csv("Netflix_X_test.csv")
y_test = pd.read_csv("Netflix_y_test.csv").squeeze()

# Encode cluster labels as discrete integers
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)

# ------------------------------------------------
# Define Base Model
# ------------------------------------------------
log_reg = LogisticRegression(max_iter=1000, random_state=42, solver='saga', multi_class='multinomial')

# ------------------------------------------------
# Define Hyperparameter Grid for Tuning
# ------------------------------------------------
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],          # Regularization strength
    'penalty': ['l1', 'l2'],                      # L1 (Lasso) and L2 (Ridge) penalties
    'solver': ['saga'],                           # saga supports both l1 and l2
}

# ------------------------------------------------
# GridSearchCV for Hyperparameter Tuning
# ------------------------------------------------
grid_search = GridSearchCV(
    estimator=log_reg,
    param_grid=param_grid,
    cv=5,                   # 5-Fold Cross Validation
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

# Fit the Algorithm (with best parameters)
grid_search.fit(X_train, y_train_encoded)

# Best parameters and CV score
print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Score:", round(grid_search.best_score_, 4))

# ------------------------------------------------
# Predict on the Model
# ------------------------------------------------
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# ------------------------------------------------
# Evaluate Model Performance
# ------------------------------------------------
test_acc = accuracy_score(y_test_encoded, y_pred)
print("\nTest Accuracy:", round(test_acc, 4))
print("\nClassification Report:\n", classification_report(y_test_encoded, y_pred))

# Confusion Matrix Visualization
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test_encoded, y_pred), annot=True, fmt='d', cmap='Greens')
plt.title('Logistic Regression (Tuned) - Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
````

**Meaning:**
Precision measures how many of the titles that were predicted to belong to a particular cluster actually belong there.

### 3. Explain the model which you have used and the feature importance using any model explainability tool?

For my final model, I used the **XGBoost Classifier**, which stands for **Extreme Gradient Boosting**.
It’s an advanced **ensemble-based boosting algorithm** that builds multiple weak decision trees sequentially — each tree focuses on correcting the errors made by the previous ones.
This approach improves both **accuracy** and **generalization**, making XGBoost one of the most efficient algorithms for structured data.

---

### **Model Explanation**

XGBoost works by:

1. Creating an initial simple model.
2. Calculating the errors (residuals).
3. Building additional trees that learn to correct those errors.
4. Combining all the trees to produce a strong final prediction model.

Each tree adds a small contribution to minimize the overall error, and **regularization parameters** (`lambda` and `alpha`) prevent overfitting — giving XGBoost a perfect balance between bias and variance.

In this Netflix ML project, XGBoost was used to classify content clusters based on various numerical and derived features such as:

* `release_year`
* `duration_minutes`
* `Title_Age`
* `Num_Genres`
* `Lexical_Richness`
* `PCA components` (reduced text features)

---

### **Feature Importance (Model Explainability)**

To understand **how the model makes its predictions**, I used XGBoost’s built-in **feature importance tool**, which ranks features based on how much they contribute to reducing model error (information gain).

The **feature importance chart** revealed that:

* **Title_Age** and **release_year** were among the top contributors — meaning the age of the content strongly influences its cluster grouping (modern vs. classic).
* **Duration_minutes** and **Num_Genres** also had high importance — longer multi-genre titles often form unique clusters.
* **Lexical_Richness** and **PCA_1/PCA_2 components** (from TF-IDF text embeddings) had significant impact — capturing content tone and complexity.

These insights help Netflix understand **which content features drive grouping behavior**, improving recommendation and categorization strategies.
